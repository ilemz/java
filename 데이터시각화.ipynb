```python
#2021년 5대광역시 출생아수와 혼인건수

import matplotlib.pyplot as plt

도시 = ['부산', '대구','인천','광주','대전']
출생아수 = [14446, 10661, 14947, 7956, 7414]
혼인건수 = [11081, 7287, 10881, 4901, 5419]

plt.plot(도시, 출생아수, label='출생아수')
plt.plot(도시, 혼인건수, label='혼인건수', linestyle='--', marker='o', color='r')
plt.legend()

plt.title('5대 광역시 혼인건수와 출생아수')
plt.xlabel('도시')
plt.ylabel('혼인(건수) 및 출생아(수)')
plt.show()
```
